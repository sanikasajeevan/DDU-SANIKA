### Importing libraries

```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

### Reading data frame

```python
df=pd.read_csv("netflix_titles_2021.csv")
```

### Assign

```python
df.assign(after_2020=df['release_year']>2020)
```

### Querry

```python
df.query('release_year==2020 & type=="Movie" & country=="India"')
```

### Dropna

```python
df_null_country=df.dropna(subset=['country'])
df_null_country
```

### Drop

```python
df1=df.drop(columns=['cast','description'])
df1
```

### Pivot table

```python
pivot=pd.pivot_table(columns='rating',values='release_year',data=df,aggfunc='median')
pivot
```
