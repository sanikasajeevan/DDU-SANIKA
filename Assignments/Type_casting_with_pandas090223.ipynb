## Type casting with pandas

```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Car_sales.csv')
df.sample(10)
```

```python
df.dropna(inplace=True)
df.info()
```

### Checking the contents in latest launch column

```python
df.Latest_Launch.unique()
```

### change its dtype into datetime

```python
df['Latest_Launch']=pd.to_datetime(df.Latest_Launch)
df.head(10)
```

```python
df.info()
```

### Extracting year,month,day from the Latest_launch column

```python
df['Year']=pd.DatetimeIndex(df.Latest_Launch).year
df['Month']=pd.DatetimeIndex(df.Latest_Launch).month
df['Day']=pd.DatetimeIndex(df.Latest_Launch).day
df.head(10)
```

### Lets convert the data type of Fuel_efficiency into integer

```python
df.Fuel_efficiency=df.Fuel_efficiency.astype(int)
df.info()
```

### converting numerical column to categorical column
##### lets convert the month numbers to their corresponding names

```python
df.Month.replace({
    1:"January",
    2:"February",
    3:"March",
    4:"April",
    5:"May",
    6:"June",
    7:"July",
    8:"August",
    9:"September",
    10:"October",
    11:"November",
    12:"December"
}, inplace=True)
df.sample(100)
```
