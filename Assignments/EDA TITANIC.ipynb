### UNIVARIATE ANALYSIS

#### KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=df.PassengerId)
plt.show()
```

#### HIST PLOT

```python
sns.histplot(df.Fare)
plt.show()
```

#### BOX PLOT

```python
sns.boxplot('Age',data=df)
plt.show()
```

### MULTI VARIATE ANALYSIS

#### LINE PLOT

```python
sns.lineplot('Age','Fare',data=df)
plt.title("Age vs Fare")
plt.show()
```

#### PIE CHART

```python
df.columns
```

```python
Pclass_Survived=df.groupby(['Pclass'])['Survived'].sum()
Pclass_Survived
```

```python
sns.set_style('ticks')
Pclass_Survived.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_Sex_Survived=df.groupby(['Pclass','Sex'])['Survived'].sum()
Pclass_Sex_Survived.reset_index()
```

```python
Pclass_Sex_Survived.plot.pie(autopct='%1.3f%%')
plt.legend(bbox_to_anchor=(1.5,1), loc='upper left',borderaxespad=1)
plt.show()
```

#### BAR CHART

```python
plt.figure(figsize=(4,3))
sns.countplot('Sex',hue='Survived',data=df)
plt.show()
```

#### COUNT PLOT

```python
plt.figure(figsize=(4,3))
sns.countplot('Survived',data=df)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Survived',data=df)
plt.title('Pclass:Survived vs Dead')
plt.show()
```

#### FACTOR PLOT

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived',hue='Sex',data=df)
plt.show()
```

#### VIOLIN PLOT

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot('Sex','Age',hue='Survived',data=df)
plt.show()
```

#### BAR PLOT

```python
plt.figure(figsize=(8,4))
sns.barplot('SibSp','Survived',data=df)
plt.title("SibSp & Survived")
plt.show()
```

#### PAIR PLOT

```python
sns.pairplot(data=df)
plt.show()
```

#### HEAT MAP

```python
heat_map=df.corr()
sns.heatmap(heat_map)
plt.show()
```
