## pd.cut method
#### it is used to seperate the array elements into different bins

### creating bins for ages

```python
df['age_bins']=pd.cut(df.Age,bins=3,labels=('young','middle','old'))
df.head(10)
```

```python
df['age_bins']=pd.cut(df.Age,bins=4,labels=('child','young','middle','old'))
df.head(10)
```

### number of age category in each class

```python
age_class=df.groupby(['Class'])['age_bins']
age_class.value_counts()
```

### count of each age category from each boarding station

```python
age_boarding=df.groupby(['Boarded'])['age_bins']
age_boarding.value_counts()
```

### count of gender and age class from each boarding station

```python
age_boarding_sex=df.groupby(['Boarded','Sex'], as_index=False)['age_bins']
age_boarding_sex.value_counts()
```
