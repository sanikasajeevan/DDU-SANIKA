```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from pydataset import data
```

```python
df=data('cake')
df
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=df.replicate)
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.countplot('temperature',hue='recipe',data=df)
plt.show()
```

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot('recipe','temperature',data=df)
plt.show()
```

```python
plt.figure(figsize=(2,4))
sns.barplot('recipe','temp',data=df)
plt.title("recipe & temprature")
plt.show()
```
